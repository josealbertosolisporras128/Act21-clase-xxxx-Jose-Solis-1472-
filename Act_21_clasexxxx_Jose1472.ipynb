```python
class Modelo:
    def __init__(self, nombre, telefono, edad):
        # Estos son los atributos
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Estas son las funciones (métodos)
    def desfilar(self):
        return f"{self.nombre} está caminando con elegancia por la pasarela."

    def promover(self):
        return f"{self.nombre} está promocionando una nueva marca en redes sociales."

    def comercializar(self):
        return f"{self.nombre} está realizando una sesión de fotos para un comercial."

# Creación de los 2 objetos (instancias de la clase Modelo)
modelo1 = Modelo("Sofía", "555-0101", 24)
modelo2 = Modelo("Alejandro", "555-0202", 28)

# Uso de los objetos y sus funciones
print(f"--- Información de {modelo1.nombre} ---")
print(f"Teléfono: {modelo1.telefono}, Edad: {modelo1.edad}")
print(modelo1.desfilar())
print(modelo1.promover())
print(modelo1.comercializar())

print("\n" + "="*30 + "\n")

print(f"--- Información de {modelo2.nombre} ---")
print(f"Teléfono: {modelo2.telefono}, Edad: {modelo2.edad}")
print(modelo2.desfilar())
print(modelo2.promover())
print(modelo2.comercializar())

print("programa terminado")
print("programa realizado por jose solis NL 51")
```
